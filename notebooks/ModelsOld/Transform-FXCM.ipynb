```python
import pandas as pd
import os
import numpy as np
from datetime import datetime
```

```python
fxcm_dir = "../data/FXCM/Raw/"
fxcm_output_dir = "../data/FXCM/Cleaned/"
```

```python
instruments = os.listdir(fxcm_dir)
```

```python
for instrument in instruments:
    output_file = fxcm_output_dir + instrument + "/" + instrument + "1.csv"
    if True: #not os.path.exists(output_file):
        
        print("Loading: {}".format(instrument))
        digits = 3 if "JPY" in instrument else 5
        
        df = pd.read_csv(fxcm_dir + instrument + "/" + instrument + "1", index_col='date')
        merged_df = pd.DataFrame()
        print("Processing: {}".format(instrument))
        merged_df["Date"] = df.index.to_frame()['date'].apply(lambda date: date.split(" ")[0]).apply(lambda date: datetime.strptime(date,"%Y-%m-%d").strftime("%Y.%m.%d"))
        merged_df["Time"] = df.index.to_frame()['date'].apply(lambda date: date.split(" ")[1]).apply(lambda time: datetime.strptime(time,"%H:%M:%S").strftime("%H:%M"))
        merged_df["Open"] = df[['bidopen', 'askopen']].mean(axis=1).round(digits)
        merged_df["High"] = df[['bidhigh', 'askhigh']].mean(axis=1).round(digits)
        merged_df["Low"] = df[['bidlow', 'asklow']].mean(axis=1).round(digits)
        merged_df["Close"] = df[['bidclose', 'askclose']].mean(axis=1).round(digits)
        merged_df["Volume"] = df["tickqty"]
        print("Dumping: {}".format(instrument))
        if not os.path.exists(fxcm_output_dir + instrument):
            os.makedirs(fxcm_output_dir + instrument)
        merged_df.to_csv(output_file, index=False, header=False)
    else:
        print("{} file exists".format(instrument))
```
