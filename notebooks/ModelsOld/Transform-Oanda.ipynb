```python
import pandas as pd
import os
import numpy as np
from datetime import datetime
from datetime import timedelta
```

```python
oanda_dir = "../data/Oanda/Raw/"
instruments = os.listdir(oanda_dir)
timeframes = ["M1", "M5", "M15", "M30", "H1", "H4", "D"]
print(instruments)
```

```python
def timeframe_to_name(timeframe):
    if timeframe == "M1":
        return "1"
    elif timeframe == "M5":
        return "5"
    elif timeframe == "M15":
        return "15"
    elif timeframe == "M30":
        return "30"
    elif timeframe == "H1":
        return "60"
    elif timeframe == "H4":
        return "240"
    elif timeframe == "D":
        return "1440"
```

```python
oanda_output_dir = "../data/Oanda/Cleaned/"


for instrument in instruments:
    for timeframe in timeframes:
        output_dir = oanda_output_dir + instrument + "/"
        output_file = output_dir + instrument + timeframe_to_name(timeframe) + ".csv"
        #if os.path.exists(output_file):
         #   print("[{} : {}] Exists, skipping".format(instrument, timeframe))
          #  continue
        print("Processing [{} : {}]".format(instrument, timeframe))
        df = pd.read_csv(oanda_dir + instrument + "/" + instrument + timeframe, index_col='Datetime')

        converted_df = df[['Datetime.1', 'Open', 'High', 'Low', 'Close', 'Volume']]
        converted_df = converted_df.rename(columns={"Datetime.1" : "Datetime", "Open" : "Open", "High" : "High", "Low" : "Low", "Close": "Close", "Volume" : "Volume"})
        digits = 3 if "JPY" in instrument else 5
        converted_df = converted_df.round({"Open" : digits, "High" : digits, "Low" : digits, "Close": digits})

        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        print("Dumping [{} : {}]".format(instrument, timeframe))
        converted_df.to_csv(output_file, index=False)
```
